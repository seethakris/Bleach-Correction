## User Input

```python
#Foldername that contains data
FolderName = '/Users/seetha/Desktop/Bleach_Correction/190116/Tiff_stack/17dpf1i/'

#Stimulus on and off time
stimulus_on_time = [10]
stimulus_off_time = [20]
crop_frame = 30
stimulus_train = ['Blue']

#Baseline over which to normalize
time_baseline = [1, 2]

#Region to analyze
region_to_analyze = 'Hb'
```

## Algorithm

### 1. Import Libraries and define functions to start thunder

```python
# Start Thunder and get thunder constant
print 'Starting Thunder Now. Check console for details'
from thunder import ThunderContext

tsc = ThunderContext.start(appName="thunderpca")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from thunder import Colorize
import os
from matplotlib.colors import ListedColormap
from matplotlib.backends.backend_pdf import PdfPages
```

```python
filesep = os.path.sep
%matplotlib inline
sns.set_context('notebook', font_scale=1.5)
image = Colorize.image
```

```python
#Create directory for Figures
Figure_PDFDirectory = FolderName + filesep + 'Figures' + filesep
if not os.path.exists(Figure_PDFDirectory):
    os.makedirs(Figure_PDFDirectory)
FishName =  os.path.basename(os.path.normpath(FolderName)) #Get fishname to append to Figures
```

### 2.Preprocess data

```python
pp = PdfPages(Figure_PDFDirectory + FishName+'_Preprocessed_Data.pdf')
```

```python
#Load functions for preprocessing first
import functions_for_preprocessing
reload(functions_for_preprocessing)
from functions_for_preprocessing import class_preprocess_data
analyze = class_preprocess_data(time_baseline, stimulus_on_time, stimulus_off_time, stimulus_train)
```

```python
filename = os.path.join(FolderName) + filesep

#Raise error if no such folder exists
if not(os.path.isdir(filename)):
    raise(ValueError('There are no such folders'))
```

```python
# Crop Image
crop = 0 
img_size_crop_y1 = 20  # How many pixels to crop on x and y axis. If none say 0
img_size_crop_y2 = 20
img_size_crop_x1 = 20  
img_size_crop_x2 = 20
```

```python
# LOad data   
data = analyze.load_and_preprocess_data(tsc, filename, crop, img_size_crop_x1, img_size_crop_x2, img_size_crop_y1,
                                 img_size_crop_y2, start_frame=crop_frame)
```

```python
#Get image of data
img_raw = data.pack()
```

```python
time_experiment = np.size(img_raw,0)
print 'Time points in experiment..', time_experiment
```

```python
# Get a small subset of data to check
examples = analyze.get_small_subset_for_plotting(data, number_samples=100, threshold=1)
```

```python
# Plot mean and traces of data and check whether to detrend or not
fig1 = plt.figure(figsize=(15,8))
if crop:
    title = 'Cropped Data :'
    txtstr = 'Cropped Data dimensions' + str(np.shape(img_raw) [1:3]) 

    ax1 = fig1.add_subplot(2,2,1)
    plt.axis('off')
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax1.text(0, 2, txtstr, transform=ax1.transAxes, fontsize=20,
            verticalalignment='top', bbox=props)
else:
    title = 'Raw Data :'

analyze.plotimageplanes(fig1, np.mean(img_raw,0), plot_title= title + region_to_analyze, num_subplots=1)
analyze.plot_traces(fig1, examples, num_subplots=2)
plt.tight_layout()
plt.show()
pp.savefig(fig1, bbox_inches='tight')
```

```python
#Detrend if necessary
detrend = 0
detrendorder = 10

fig1 = plt.figure(figsize=(15,8))
if detrend:
    data = analyze.detrend_data(data, detrend_order=detrendorder)
    data.cache()
    examples = analyze.get_small_subset_for_plotting(data, number_samples=100, threshold=10)
    analyze.plot_traces(fig1, examples, num_subplots=1, \
                        plot_title='Detrended Data: ' + region_to_analyze + \
                        ' detrend_order=' + str(detrendorder) )
plt.show()
pp.savefig(fig1)
```

```python
pp.close()
```

### 3. Normalize data

```python
pp = PdfPages(Figure_PDFDirectory + FishName +'_Normalized_smoothed.pdf')
```

```python
# Get zscore of data
zscore = analyze.normalize(data, squelch_parameter=20)
```

```python
# Get subset for plotting
examples = analyze.get_small_subset_for_plotting(zscore, number_samples=10, threshold=1)
```

```python
# Get image of zscore
img = zscore.pack()
```

```python
fig1 = plt.figure(figsize=(10,8))
analyze.plotimageplanes(fig1, np.mean(img,0),  plot_title='Normalized Data: '+region_to_analyze, num_subplots=1)
analyze.plot_traces(fig1, examples, num_subplots=2)
plt.suptitle('Time_Baseline='+str(time_baseline), fontsize=20)
plt.tight_layout()
plt.show()
pp.savefig(fig1, bbox_inches='tight')
```

```python
# Smooth data if necessary and plot
smooth = 1
windowlen = 10

if smooth:
    zscore_smooth = data.applyValues(lambda x: analyze.smooth_func(x, window_len=windowlen), keepIndex=True)
    zscore_smooth.cache()
    examples = analyze.get_small_subset_for_plotting(zscore_smooth, number_samples=50, threshold=1)  
```

```python
if smooth:
    
    fig1 = plt.figure(figsize=(15,8))
    analyze.plot_traces(fig1, examples, num_subplots=1, \
                       plot_title='Smoothed Data: ' + region_to_analyze + \
                        ' moving average window =' + str(windowlen))
    plt.show()

    pp.savefig(fig1)
```

### Use standard deviation to threshold data if necessary

```python
do_std = 0
if do_std:
    stdMap = zscore.seriesStdev().pack()
    
    threshold = 1
    
    fig1 = plt.figure(figsize=(15,8))
    analyze.plotimageplanes(fig1, stdMap>threshold ,\
                            plot_title='Standard_Deviation: '+ region_to_analyze + \
                            ' threshold =' + str(threshold),\
                            num_subplots=1)
    
    filtered = zscore.filterOnValues(lambda x: np.std(x) > threshold)
    
    plt.show()
    pp.savefig(fig1)
```

```python
pp.close()
```

### 4. Perform KMeans

```python
pp = PdfPages(Figure_PDFDirectory + FishName + '_KMeans.pdf')
```

```python
#Kmeans_Parameters
kmeans_clusters = 4
kmeans_data = zscore
```

```python
#Load functions for kmeans 
import functions_for_kmeans
reload(functions_for_kmeans)
from functions_for_kmeans import class_kmeans
kmeans = class_kmeans(kmeans_clusters, kmeans_data, img_raw, stimulus_on_time, stimulus_off_time, stimulus_train)
```

```python
# Perform kmeans
model, img_sim, img_labels = kmeans.run_kmeans()
```

```python
kmeans_clusters = model.centers.T
```

```python
#Check plots with a random colormap
fig1 = plt.figure(figsize=(10,10))
gs = plt.GridSpec(1, 2, width_ratios=[1,0.2])
cmapCat = ListedColormap(sns.color_palette("Set1", n_colors=8), name='from_list')
kmeans.plot_kmeans_components(fig1, gs, kmeans_clusters, cmapCat.colors,\
                                  plot_title='Kmeans_clusters : ' + region_to_analyze,
                                  num_subplots=1, gridspecs='[0,0]')
plt.tight_layout()
plt.show()
```

```python
#remove noisy clusters and replot
ignore_clusters = [0, 2];
```

```python
#Check plots with a random colormap
fig1 = plt.figure(figsize=(10,10))
gs = plt.GridSpec(2, 2, width_ratios=[1,0.2])
cmapCat = ListedColormap(sns.color_palette("Set1", n_colors=8), name='from_list')
kmeans.plot_kmeans_components(fig1, gs, kmeans_clusters, cmapCat.colors,\
                                  plot_title='Kmeans_clusters : ' + region_to_analyze,
                                  num_subplots=1, gridspecs='[0,0]')

txtstr = 'ignore_clusters_'+ region_to_analyze + ': ' + str(ignore_clusters)

ax1 = fig1.add_subplot(gs[0,1])
plt.axis('off')
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax1.text(-4, 1.2, txtstr, transform=ax1.transAxes, fontsize=20,
        verticalalignment='top', bbox=props)

plt.tight_layout()
pp.savefig(fig1,  bbox_inches='tight')
```

```python
#Use a colorbrewer colormap to create map of brain
brainmap, unique_clrs, newclrs_updated_rgb, newclrs_updated_brewer, matched_pixels, \
kmeans_clusters_updated = kmeans.make_kmeans_maps(kmeans_clusters, \
                                                          img_labels, img_sim, mixing_parameter=0.8,
                                                          std_threshold=0.0005, ignore_clusters=ignore_clusters)
```

```python
fig1 = plt.figure(figsize=(20,5))
gs = plt.GridSpec(1,5,width_ratios=[5,0.1,3,0.5,1])
kmeans.plotimageplanes(fig1, gs, brainmap , plot_title='Brain Map : '+region_to_analyze, gridspecs='[0,0]')
kmeans.plot_kmeans_components(fig1, gs, kmeans_clusters_updated, newclrs_updated_brewer.colors,\
                                  plot_title='Kmeans_clusters : ' + region_to_analyze, num_subplots=2, flag_separate=0, gridspecs='[0,2]')
kmeans.plot_matchedpixels(fig1, gs, matched_pixels, unique_clrs, gridspecs='[0,4]')

plt.tight_layout()
plt.show()
pp.savefig(fig1, bbox_inches='tight')
```

```python
pp.close()
```

### Run PCA

```python
pp = PdfPages(Figure_PDFDirectory + FishName + '_PCA.pdf')
```

```python
PCA_data = data
```

```python
# PCA parameters 
pca_components = 4  # Number of pca components to detect from files
num_pca_colors = 50  # Number of colors on the pca maps
num_samples = 10000  # number of random samples to select to do PCA reconstruction
thresh_pca = 0.000001  # Threshold above which to plot the pca components
color_map = "polar"

#Color to plot each trial
# color_mat = ['#00FFFF', '#FF0000', '#0000FF', '#FF1493', '#3090C7', '#800000'] #Blue red X3
color_mat = ['#00FFFF','#FF1493', '#FF8C00', '#0000FF', '#9400D3', '#FF4500'] #Blue uv us x2
```

```python
#Load functions for PCA
import functions_for_PCA
reload(functions_for_PCA)
from functions_for_PCA import class_PCA
PCA = class_PCA(pca_components, PCA_data, img_raw, num_pca_colors, num_samples,\
                    thresh_pca, color_map, color_mat,stimulus_on_time, stimulus_off_time, stimulus_train)
```

```python
#Run PCA
required_pcs = [1,2]
```

```python
model, imgs = PCA.run_pca(required_pcs)
```

```python
maps, pts, clrs, recon, unique_clrs_PCA, matched_pixels_PCA, \
               matched_signals = PCA.make_pca_maps(model, imgs, \
                                                       required_pcs, mixing_parameter=1.0)
```

```python
pca_components = model.comps.T
pca_eigenvalues = model.latent
```

```python
fig1 = plt.figure(figsize=(10,10))

gs = plt.GridSpec(3,5, height_ratios=[0.25,0.25,1], width_ratios=[0.5,1.5,0.25, 0.5,1.5])
PCA.plot_pca_components(fig1, gs, pca_components, required_pcs, plot_title='PCA components : ' + region_to_analyze, gridspecs='[0,0:2]')

PCA.plot_eigenvalues(fig1, gs, pca_eigenvalues, gridspecs='[1, 0:2]')

PCA.plot_stimulus_in_3d(fig1, gs, pca_components, required_pcs, 'z', gridspecs='[2,0:2]')

plt.tight_layout()
plt.show()
pp.savefig(fig1, bbox_inches='tight')
```

```python
fig1 = plt.figure(figsize=(20,10))
gs = plt.GridSpec(2,5,width_ratios=[5,0.2,3,0.5,1])
PCA.plotimageplanes(fig1, gs, maps , plot_title='PCA map: ' + region_to_analyze,\
                        gridspecs='[0,0]')

PCA.plot_scores(fig1, gs, matched_signals, unique_clrs_PCA, plot_title='PCA scores : ' + region_to_analyze, gridspecs='[0,2]')

PCA.plot_matchedpixels(fig1, gs, matched_pixels_PCA, unique_clrs_PCA, gridspecs='[0,4]')

plt.tight_layout()
plt.show()
pp.savefig(fig1, bbox_inches='tight')
```

### Regression

```python
pp = PdfPages(Figure_PDFDirectory + FishName + '_Regression.pdf')
```

```python
regression_data = data
```

```python
#Load functions for regression
import functions_for_regression
reload(functions_for_regression)
from functions_for_regression import class_regression
regression = class_regression(regression_data, img_raw, time_experiment, stimulus_on_time, stimulus_off_time, stimulus_train)
```

```python
# Get regressors
amplitude_for_stimulus_train = 5
smooth_window_length = 10
fig1 = plt.figure(figsize=(5, 4))
regressors, regressorlist = regression.create_regression_parameters(fig1, amplitude_for_stimulus_train, smooth_window_length, plot_flag=1)
plt.tight_layout()
plt.show()
pp.savefig(fig1,  bbox_inches='tight')
```

```python
# Plot correlation coefficient with regressors 
fig1 = plt.figure(figsize=(15,10))
A = list(np.ones(len(regressors)))
A.append(0.15)
A = np.array(A)

gs = plt.GridSpec(len(regressors)+1, 2, height_ratios=A)

corrMat = np.zeros((np.size(img,1),np.size(img,2),len(regressors)))

count = 0
for key, value in regressors.iteritems():
    
    if count == len(regressors)-1:
        colorbar = True
    else:
        colorbar = False
    corrMat[:,:,count] = regression.plot_correlation_with_regressors(fig1, gs, key, value, gridspecs='['+ str(count)+',0]',
                                        color_bar=colorbar, region=region_to_analyze, clim=[-0.5, 0.5])

    
    count += 1

plt.suptitle('Correlation Coefficients with regressors')
plt.tight_layout()
plt.show()
pp.savefig(fig1)
```

```python
regression_results, betas, rsq = regression.perform_regression(regressorlist)
```

```python
#Plot r-squared
fig1 = plt.figure(figsize=(15,10))
gs = plt.GridSpec(1, 2)

regression.plotimageplanes(fig1, gs, rsq, cmap = 'gray', color_bar=True, plot_title='Rsquare: '+region_to_analyze, gridspecs='[0,0]')

plt.tight_layout()
plt.show()
pp.savefig(fig1)
```

```python
# plot betas seperately
fig1 = plt.figure(figsize=(20 ,5))
gs = plt.GridSpec(2, len(regressors))
for ii in xrange(0, len(regressors)):
    b1map = regression.center(betas[ii,:,:])
    regression.plotimageplanes(fig1, gs, b1map, cmap = 'seismic', color_bar=False, \
                               plot_title='Betas :' + region_to_analyze + ' ' +regressors.keys()[ii],\
                               gridspecs='[0,'+str(ii)+']', clim=[-0.5, 0.5])
    
    
    if count == len(regressors)-1:
        colorbar = True
    else:
        colorbar = False
plt.show()
pp.savefig(fig1)
```

```python
# Plot regressors as an RGB on the habenula mean image plane 
colors = ['aqua', 'green', 'orange', 'DarkRed', 'Fuchsia','DarkSlateBlue']
fig1 = plt.figure(figsize=(20 ,10))
gs = plt.GridSpec(4, len(regressors), height_ratios=[1,1,0.2,3])

regression.plot_regressors_as_RGB(fig1, gs, regressors, rsq, betas, colors, brightness_scale=10, mixing_parameter=0.5)

plt.show()
pp.savefig(fig1)
```
